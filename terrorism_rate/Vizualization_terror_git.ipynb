# Визуализация количества погибших от террористических актов

```python
import pandas as pd
import numpy as np
import bar_chart_race as bcr
from IPython.display import HTML
import warnings
warnings.filterwarnings('ignore')
```

```python
url = 'https://drive.google.com/file/d/18vvS4rfELBXCUT6z5aht_LRayW3eUaFu/view?usp=sharing'
url_2 = 'https://drive.google.com/uc?id=' + url.split('/')[-2]
```

```python
df = pd.read_csv(url_2, parse_dates=[0])
```

```python
df.head()
```

```python
df.info()
```

```python
df = df.fillna(0)
```

```python
df = df.rename(columns={'Terrorism fatalities (GTD, 2018)':'terrorism_rate'})
```

```python
df.info()
```

```python
df['Year'].describe()
```

```python
df = df[df['Year'] >= 2000]
```

```python
df['Year'] = pd.to_datetime(df['Year'], format='%Y')
```

```python
df['Entity'].unique()
```

```python
deleted_locations=['World', 'South America', 'South Asia', 'Southeast Asia', 'Sub-Saharan Africa', 'North America', 
                   'Middle East & North Africa', 'International', 'East Asia', 'Eastern Europe', 'Central America & Caribbean', 
                  'Central Asia', 'Australasia & Oceania', 'Yugoslavia']
```

```python
df = df[~df['Entity'].isin(deleted_locations)]
```

## Построенине графика race

```python
df_terror = df.pivot_table(index='Year', columns='Entity', values='terrorism_rate', aggfunc='sum')
```

```python
df_terror.head()
```

```python
df_terror = df_terror.cumsum(axis=0)
```

```python
df_terror.head()
```

```python
bcr.bar_chart_race(
    df_terror,
    filename=None,
    orientation='h',
    sort='desc',
    n_bars=15,
    fixed_order=False,
    fixed_max=True,
    steps_per_period=5,
    interpolate_period=False,
    label_bars=True,
    bar_size=.5,
    period_label={'x': .99, 'y': .25, 'ha': 'right', 'va': 'center', 'size':17, 'weight':'bold'},
    period_fmt='%Y',
    period_summary_func=lambda v, r: {'x': .99, 'y': .15,
                                      's': f'Погибших с 2000 года: {v.sum():,.0f}',
                                      'ha': 'right', 'size': 10, 'family': 'sans-serif', 'weight':'bold'},
    period_length=500,
    figsize=(5, 5),
    cmap='Dark2',
    title='   ',
    title_size=12,
    bar_label_size=7,
    tick_label_size=7,
    shared_fontdict={'family':'sans-serif', 'color':'#000000', 'weight':'normal'},
    scale='linear',
    writer=None,
    fig=None,
    bar_kwargs={'alpha': .8},
    filter_column_colors=False)
```
